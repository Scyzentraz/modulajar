```python
#img2img
# Step 1: Upload file
from google.colab import files
uploaded = files.upload()

# Ambil nama file yang baru diupload
import io
from PIL import Image

filename = list(uploaded.keys())[0]
input_image = Image.open(io.BytesIO(uploaded[filename])).convert("RGB")

# Pastikan resolusi adalah kelipatan 8
def resize_to_multiple_of_8(img):
    width, height = img.size
    new_width = (width // 8) * 8
    new_height = (height // 8) * 8
    return img.resize((new_width, new_height), Image.LANCZOS)

input_image = resize_to_multiple_of_8(input_image)

# Step 2: Setup pipeline
from diffusers import StableDiffusionImg2ImgPipeline
import torch

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

# Step 3: Denoise image with low strength
prompt = (
    "Ultra high-resolution early morning forest landscape, clean natural lighting with soft morning mist, "
    "realistic tree branches and leaves with sharp organic structure, smooth sky gradients, natural sunlight rays, "
    "well-defined foliage, high clarity and realistic color grading, sharp focus, noise-free, photorealistic, "
    "true-to-life tones, suitable for stock photography"
)

negative_prompt = (
    "oil painting, blurry, artifact, glitch, oversaturated, noise, deformed details, brush strokes, "
    "fake glow, overprocessed, CG look, watercolor, distortion, soft edges, jpeg artifacts, halo, smudge, posterization"
)

result = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    image=input_image,
    strength = 0.25,
    guidance_scale = 8.0,
    num_inference_steps = 40)

# Step 4: Save output
output_path = "denoised_output.png"
result.images[0].save(output_path)

# Tampilkan hasil
from IPython.display import Image as IPyImage
IPyImage(output_path)
```
